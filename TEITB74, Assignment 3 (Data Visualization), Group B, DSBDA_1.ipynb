```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("airquality1.csv")
```

```python
data
```

```python
data["Ozone"] = data["Ozone"].fillna(data["Ozone"].mean())
data["Solar.R"] = data["Solar.R"].fillna(data["Solar.R"].mean())
```

```python
data
```

```python
plt.pie([data["Ozone"].mean(),data["Solar.R"].mean(),data["Wind"].mean(),data["Temp"].mean()], autopct="%1.2f%%", explode=[0.01,0.01,0.01,0.01], labels=["Ozone", "Solar.R", "Wind", "Temp"])
plt.plot()
```

```python
def addlabels():
    for i in range(len(data.iloc[0:31, 4])):
        plt.text(i, data.iloc[0:31, 4][i], data.iloc[0:31, 4][i], ha = 'center')
        
y = np.arange(len(data.iloc[0:31, 4]))
addlabels()
plt.bar(x=y, height=data.iloc[0:31, 4], tick_label=y, color="dodgerblue")
plt.title("Temeprature for 1st to 30th of May")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.show()
```

```python
plt.boxplot(x=data[["Temp", "Solar.R", "Wind", "Ozone"]], labels=["Temp", "Solar.R", "Wind", "Ozone"])
plt.plot()
```

```python
plt.hist(x=data["Temp"], bins=20, color="orange", label="Temp")
plt.title("Temperature count")
plt.xlabel("Temperature")
plt.ylabel("Count")
plt.plot()
```

```python
plt.plot(data.iloc[0:30, 6], data.iloc[0:30, 3], color="purple")
plt.title("Wind of May month")
plt.xlabel("Day")
plt.ylabel("Wind")
```

```python
plt.scatter(data.iloc[0:30, 6], data.iloc[0:30, 3], color="red")
plt.title("Wind measure of May month")
plt.xlabel("Day")
plt.ylabel("Wind")
```
